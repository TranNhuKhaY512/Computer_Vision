```python
import cv2
import matplotlib.pyplot as TranNhuKhaY
import numpy as np
```

## 1. Đọc ảnh bằng OpenCV

```python
import cv2

img = cv2.imread("/content/drive/MyDrive/CV_2024_data/picture.png")

print(type(img))
print(img.shape)
```

## 2. Hiển thị ảnh bằng plt.imshow & nhận xét

```python


TranNhuKhaY.imshow(img)
TranNhuKhaY.title("Image using plt.imshow")
TranNhuKhaY.axis("off")
TranNhuKhaY.show()
```

```python
# Ảnh bị ám màu xanh / sai màu
# Nguyên nhân: plt.imshow() hiểu ảnh là RGB, trong khi OpenCV đọc ảnh là BGR
```

### Cách hiển thị đúng màu

```python
# cách 1 đảo chiều ma trix
img_rgb = img[:, :, ::-1]

TranNhuKhaY.imshow(img_rgb)
TranNhuKhaY.axis("off")
TranNhuKhaY.show()
```

```python
# cách 2 convert chế độ màu
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

TranNhuKhaY.imshow(img_rgb)
TranNhuKhaY.axis("off")
TranNhuKhaY.show()
```

## 4. Phóng to ảnh (Resize)

```python

h, w = img.shape[:2]

img_resize = cv2.resize(img, (w*2, h*2))

TranNhuKhaY.imshow(cv2.cvtColor(img_resize, cv2.COLOR_BGR2RGB))
TranNhuKhaY.axis("off")
TranNhuKhaY.show()
```

## Crop ảnh

```python
# Crop bằng slicing NumPy
# Cắt ảnh: y1:y2, x1:x2
crop_img = img[100:300, 150:400]

TranNhuKhaY.imshow(cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB))
TranNhuKhaY.axis("off")
TranNhuKhaY.show()
```

## Ghép 2 ảnh (ngang & dọc)

```python
img1 = cv2.imread("/content/drive/MyDrive/CV_2024_data/p1.png")
img2 = img1.copy()
```

```python


# ghép ngang
import numpy as np

horizontal = np.hstack((img1, img2))

TranNhuKhaY.imshow(cv2.cvtColor(horizontal, cv2.COLOR_BGR2RGB))
TranNhuKhaY.axis("off")
TranNhuKhaY.show()
```

```python
# ghép dọc
vertical = np.vstack((img1, img2))

TranNhuKhaY.imshow(cv2.cvtColor(vertical, cv2.COLOR_BGR2RGB))
TranNhuKhaY.axis("off")
TranNhuKhaY.show()
```
